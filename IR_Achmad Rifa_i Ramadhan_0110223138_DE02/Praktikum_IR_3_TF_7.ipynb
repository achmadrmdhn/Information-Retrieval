```python
import re  # Mengimpor modul re untuk manipulasi string menggunakan ekspresi reguler
from collections import Counter  # Mengimpor Counter dari modul collections untuk menghitung frekuensi kata
```

```python
# Membaca file teks
with open('document.txt.txt', 'r', encoding='utf-8') as file:  # Membuka file 'document.txt.txt' dalam mode baca dengan encoding UTF-8
    text = file.read()  # Membaca seluruh isi file ke dalam variabel 'text'
```

```python
 # Mengubah teks menjadi huruf kecil dan menghapus tanda baca
text = text.lower()  # Mengubah semua teks menjadi huruf kecil agar pencocokan kata tidak peka terhadap kapitalisasi
text = re.sub(r'[^\w\s]', '', text)  # Menghapus semua karakter non-alfabet dan non-spasi (tanda baca) menggunakan ekspresi reguler
```

```python
# Tokenisasi - memecah teks menjadi kata-kata
words = text.split()  # Memecah teks yang sudah diproses menjadi daftar kata-kata menggunakan spasi sebagai pemisah
```

```python
words
```

```python
# Menghitung frekuensi setiap kata dalam teks
word_counts = Counter(words)  # Menghitung jumlah kemunculan setiap kata dalam daftar 'words' menggunakan Counter
```

```python
word_counts
```

```python
 # Menghitung total kata dalam dokumen
total_words = sum(word_counts.values())  # Menjumlahkan semua frekuensi kata untuk mendapatkan jumlah total kata
```

```python
total_words
```

```python
# Menghitung Term Frequency (TF) untuk setiap kata
tf_scores = {word: count / total_words for word, count in word_counts.items()}  # Menghitung Term Frequency (TF) setiap kata
```

```python
tf_scores
```

```python
# Menampilkan hasil TF setiap kata
print("Hasil Term Frequency (TF):")  # Menampilkan header "Hasil Term Frequency (TF):"
for word, tf in tf_scores.items():  # Iterasi melalui semua kata dan skor TF
    print(f"{word}: {tf:.4f}")  # Menampilkan kata dan skor TF-nya dengan format 4 desimal
```
