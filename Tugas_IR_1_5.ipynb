```python
import nltk
nltk.download('punkt') #untuk tokenisasi
nltk.download('stopwords') #untuk stopword
nltk.download('wordnet') #untuk lemmatization
```

```python
from nltk.tokenize import sent_tokenize #Memotong paragraf menjadi kalimat

text="""
Editing video adalah salah satu elemen penting dalam produksi media yang melibatkan proses menyusun, memotong, dan menggabungkan klip video untuk menghasilkan narasi yang koheren dan menarik. Proses ini tidak hanya sekedar merangkai potongan video, namun juga melibatkan pemahaman yang mendalam tentang ritme, estetika visual, dan teknik penyuntingan agar hasil akhir bisa menyampaikan cerita dengan kuat dan memukau penonton.
"""
```

```python
# Tokenisasi menjadi kalimat
senteces = sent_tokenize(text)
senteces
```

```python
# Tokenisasi menjadi kata-kata
from nltk.tokenize import word_tokenize
words = word_tokenize(text)
print(words)
```

```python
# Menghilangkan stop words dan tanda baca
from nltk.corpus import stopwords
word_in_text= word_tokenize(text)
print(word_in_text)
stop_words = set(stopwords.words('indonesian'))
print(stop_words)
#inisialisasi apa yang mau dibuang
stop_words.add(',')
stop_words.add('.')
stop_words.add('(')
stop_words.add(')')
stop_words.add('-')
stop_words.add('"')
```

```python
#filter kata untuk menghilangkan stop words dan tanda baca
filtered_words = [word for word in words if word.lower() not in stop_words]
filtered_words
```

```python
# Stemming
from nltk.stem import PorterStemmer

ps = PorterStemmer()
stemmed_words = [ps.stem(word) for word in filtered_words]
stemmed_words
```

```python
# Lemmatization
from nltk.stem import WordNetLemmatizer

lemmatizer= WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]
print(lemmatized_words)
```

```python
from collections import Counter
import matplotlib.pyplot as plt

# Menghitung frekuensi kata
word_freq = Counter(lemmatized_words)
most_common_words = word_freq.most_common(20)


# Visualisasi frekuensi kata
words, counts= zip(*most_common_words)
plt.bar(words, counts)
plt.xlabel('Kata-Kata')
plt.ylabel('Frekuensi')
plt.xticks(rotation=45)
plt.title('Kata-Kata yang Paling Umum dalam Artikel Editing Video')
plt.show()
```
